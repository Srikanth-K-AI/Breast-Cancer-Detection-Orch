**Converting Diagnosis column to Numerical value**

```python
dataset['diagnosis'] =  dataset['diagnosis'].map({'M': 1 , 'B' : 0})
dataset
```

**Correlation with Target feature**

```python
correlation = dataset.corr()
correlation
```

```python
plt.figure(figsize =(28,15))
sns.heatmap(correlation , annot = True , cmap='coolwarm')
plt.title("Correlation matrix")
```

```python
# correlation between indepandent variables with target value
target_corr = correlation['diagnosis'].sort_values(ascending = True)
target_corr
```

**Feature Removel**
- Based on the Correlation analysis , the following features showed very weak relationship with target feature . and drop to simplify the model
- ID (correlation == 0.04)
- fractal_dimension_mean (Correlation = -0.013)
- texture_se (correlation = -0.0083)
- smothness_se (correlation = -0.067)
- symmetry_se (correlation = -0.0065)

```python
# drop features
drop_colums = ['id','fractal_dimension_mean','texture_se','smoothness_se','symmetry_se']
dataset.drop(columns = drop_colums , inplace=True)
```

```python
dataset
```

**Seperating the features and Target**

```python
y = dataset.iloc[: , 0]   #features / independent features
x = dataset.iloc[ : , 1:-1]  # target / dependent feature
```

**Split the data into training and testing data**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size = 0.2 , random_state = 2)
print(x.shape , x_train.shape , y_train.shape)
print(x.shape , x_test.shape , y_test.shape)

```

**Feature Scalling**
- Applying StandardScaler 

```python
#from sklearn.preprocessing import standardscaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

**Model Training**
- Logistic Regression

```python
model = LogisticRegression(max_iter=10000)
model.fit(x_train,y_train)
```

**Model evaluation**

```python
# from sklearn.metrics import accuracy_score

#training data accuracy
x_train_prediction = model.predict(x_train)
accuracy_of_training_data = accuracy_score(y_train,x_train_prediction)*100
print(f"Training data accuracy : {accuracy_of_training_data}")
```

```python
#testing data accuracy
x_test_prediction = model.predict(x_test)
accuracy_of_test_data = accuracy_score(y_test , x_test_prediction)*100
print(f"Test data accuracy : {accuracy_of_test_data} ")
```

```python
#Check Precision , recall ,
print(classification_report(y_test , x_test_prediction))
```

**Save the Model**

```python
 

# Save the model to a file
with open('breast_cancer_model.pkl', 'wb') as file:
    pickle.dump(model, file)

#SAVE THE SCALER 
with open('scaler.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)
```
